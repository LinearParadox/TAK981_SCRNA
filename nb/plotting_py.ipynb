```python
import scanpy as sc
import matplotlib.pyplot as plt
import os
from glob import glob
import pandas as pd
os.chdir("/data/TAK_981/")
```

```python
adata = sc.read_h5ad("/data/TAK_981/h5ad/batch_corrected.h5ad")
adata = adata.raw.to_adata()
```

```python
sc.pp.normalize_total(adata)
sc.pp.log1p(adata)
```

```python
ep = adata[adata.obs.annot == "Epithelial"]
fig, [ax1, ax2, ax3, ax4] = plt.subplots(1,4)
fig.set_figheight(4)
fig.set_figwidth(12)
sc.pl.violin(ep[ep.obs.condition=="ctrl"], keys="H2-K1", use_raw=False, ax=ax1, show=False, size=0)
sc.pl.violin(ep[ep.obs.condition=="tak"], keys="H2-K1", use_raw=False, ax=ax2, show=False, size=0)
sc.pl.violin(ep[ep.obs.condition=="cd40"], keys="H2-K1", use_raw=False, ax=ax3, show=False, size=0)
sc.pl.violin(ep[ep.obs.condition=="combo"], keys="H2-K1", use_raw=False, ax=ax4, show=False, size=0)
for ax in [ax1, ax2, ax3, ax4]:
    ax.set_ylim(0, 6)
ax1.set_title("CTRL MHCI")
ax2.set_title("TAK MHCI")
ax3.set_title("CD40 MHCI")
ax4.set_title("COMBO MHCI")
plt.tight_layout()
plt.savefig("figs/MHC_I_violin.png")
plt.show()
```

Creating DF For p values for these comparisons

```python
genes = ("H2-K1", "Tgfb1", "Tgfb2", "Tgfb3", "H2-D1", "H2-T23")
files = glob("csv/de/memento/Epithelial*")
dfs = [pd.read_csv(n, index_col="Unnamed: 0") for n in files]
subset_df = []
for file, df in zip(files,  dfs):
    name = file.split("/")[-1][11:-4]
    filt = df.loc[df["gene"].isin(genes)]
    filt.loc[:, "tx"] = name
    subset_df.append(filt)
pd.concat(subset_df).to_csv("csv/de/genes_of_interest.csv", index=False)   
```
